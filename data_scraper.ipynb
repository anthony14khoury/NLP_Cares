# Wikipedia - Glossary of Medicine

```python
import wikipediaapi

page_title = "Glossary of medicine"

# Create a Wikipedia object
wiki_wiki = wikipediaapi.Wikipedia(user_agent='anthony14khoury@gmail.com')

# Fetch the page
page = wiki_wiki.page(page_title)

sections = page.sections[:-2]
```

```python
terms, definitions = [], []
missed = []

for section in sections:
        
    for s in section.text.split('\n'):
        try:
            term, definition = s.split(" – ", 1)
            
            if len(definition) == 0:
                missed.append(s)
                continue
        
            terms.append(term)
            definitions.append(definition)
        
        except:
            try:
                term, definition = s.split("-", 1)
        
                terms.append(term)
                definitions.append(definition)
            
            except:
                try:
                    term, definition = s.split("-", 1)
            
                    terms.append(term)
                    definitions.append(definition)
            
                except:
                    
                    missed.append(s)
    
print("Caught: {} | Missed: {}".format(len(terms), len(missed)))
```

```python
np.save("terms_wiki_glossary_of_medicine.npy", np.array(terms))
np.save("definitions_wiki_glossary_of_medicine.npy", np.array(definitions))
```

### Outputting Terms, Definitions, and Missed (empty)

# UMich Medical Dictionary

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import tqdm

# Create a new instance of the Chrome driver
driver = webdriver.Chrome()

# Open the website
driver.get('https://apps.lib.umich.edu/medical-dictionary/#')
```

```python
elements = driver.find_elements(By.CLASS_NAME, "term-card")
```

```python
len(elements)
```

```python
terms, definitions = [], []
i = 0
for e in elements:
    if i % 10 == 0:
        print(i)
    
    # Try to click 'expand'
    try:
        a_tag = e.find_element(By.TAG_NAME, 'a')
        a_tag.click()
    except:
        pass
    
    text = str(e.text)
    text = text.replace(" collapse", "")
    
    term, definition = text.split('\n')
    terms.append(term)
    definitions.append(definition)
    
    i += 1
```

```python
for t, d in zip(terms, definitions):
    print(t)
    print(d)
    print()
```

```python
np.save("terms_umich_medical_dictionary.npy", np.array(terms))
np.save("definitions_umich_medical_dictionary.npy", np.array(definitions))
```

# Plain Language Thesaurus

```python
import PyPDF2

terms, definitions, missed = [], [], []

# Open the PDF file
with open('cdc_plt.pdf', 'rb') as file:
    # Create a PDF file reader object
    pdf_reader = PyPDF2.PdfReader(file)

    # Loop through each page of the PDF
    for page_num in range(2, len(pdf_reader.pages)):
        
        # Get a specific page
        page = pdf_reader.pages[page_num]

        # Extract text from the page
        text = page.extract_text()
        
        text = text.split("092607")[1]
        text = text[2:] # Remove Page Number
        
        text_split = text.split("  ")
        
        # Handling first term and definition
        terms.append(text_split[0].split("\n")[-1])
        
        definition = text_split[1].split("\n")     
        if len(definition) > 1:
            definition = "".join(definition[:-1])
            definition = definition[:-1] # Removing space
            definitions.append(definition)
        else:
            definitions.append(definition[0])
        
        
        i = 2
        while i < len(text_split):        
            try:
                term = text_split[i-1]
                if term.find('\n') == -1:
                    i += 1
                    continue
                
                term = text_split[i-1].split("\n")[-1]
                
                # Error Checking
                if term.find(',') != -1 or term.find(';') != -1:
                    missed.append(text_split[i])
                    i += 1
                    continue
                
                terms.append(term)
                
                definition = text_split[i].split("\n")
                
                # Combining terms
                if len(definition) > 1:
                    definition = "".join(definition[:-1])
                    definition = definition[:-1] # Removing space
                    definitions.append(definition)
                else:
                    definitions.append(definition[0])
                
            except:
                missed.append(text_split[i])
            
            i += 1
```

### Catching any uncaught errors

```python
# Iterate through both arrays backwards by index for deletion
for i in reversed(range(len(terms))):
    term = terms[i]
    definition = definitions[i]
    
    # Deleting the elements that are empty
    if definition == '':
        del terms[i]
        del definitions[i]
        continue
    
    # Removing space in the front of the string
    if definition[0] == ' ':
        definitions[i] = definition[1:]
```
